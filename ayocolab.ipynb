# **Numeric**
1. Integer

    Nilai ini diwakili oleh kelas int. Berisi bilangan bulat positif atau negatif (tanpa pecahan atau desimal). Di Python, tidak ada batasan berapa panjang suatu nilai integer.

```python
number = 4

print(number)

print(type(number))
```

2. **Float**

    Nilai ini direpresentasikan oleh kelas float. Ini adalah bilangan riil dengan representasi floating-point. Nilai ini ditentukan oleh titik desimal. Secara opsional, karakter e atau E diikuti oleh bilangan bulat positif atau negatif dapat ditambahkan untuk menentukan notasi ilmiah.

```python
number = 5.0

print(number)

print(type(number))
```

3. **Complex**

     Bilangan kompleks direpresentasikan oleh kelas kompleks. Bilangan ini dinyatakan sebagai (bagian riil) + (bagian imajiner)j . Misalnya  2+3j

```python
number = 2 + 3j

print(number)

print(type(number))
```

# **Sequence Data Types**


Urutan Tipe Data dalam Python adalah kumpulan terurut dari tipe data Python yang serupa atau berbeda. Urutan memungkinkan penyimpanan berbagai nilai secara terorganisir dan efisien.


1. **String**

String Python adalah array byte yang mewakili karakter Unicode. Di Python, tidak ada tipe data karakter Python, karakter adalah string yang panjangnya satu. Itu diwakili oleh kelas str.

String di Python dapat dibuat menggunakan tanda kutip tunggal, tanda kutip ganda, atau bahkan tanda kutip tiga. Kita dapat mengakses karakter individual dari sebuah String menggunakan indeks.

```python
kata_kata = "Halo Dunia Coding"

print(kata_kata)

print(type(kata_kata))

print(kata_kata[0])
print(kata_kata[1])
print(kata_kata[2])
```

2. **List**

List sama seperti array, dideklarasikan dalam bahasa lain yang merupakan kumpulan data yang diurutkan. Ini sangat fleksibel karena item dalam daftar tidak harus bertipe sama.

```python
ini_list = [] #list kosong

ini_list = [1, 2.0, "tahu", "tempe", True] #list tidak kosong

print(ini_list)

# print(type(ini_list))

# print(ini_list[0])

ini_list[0] = 'Satu'

print(ini_list)
```

3. **Tuple**

Sama seperti List, tuple juga merupakan kumpulan objek Python yang diurutkan. Satu-satunya perbedaan antara tupel dan List adalah bahwa tupel tidak dapat diubah isinya / tupel tidak dapat diubah setelah dibuat.

```python
ini_tupple = () # tuple kosong

ini_tupple = (1, 2, 'Tahu', "Tempe")

print(ini_tupple)

print(type(ini_tupple))

print(ini_tupple[1])

ini_tupple[0] = 'Satu'

print(ini_tupple)
```

4. **Boolean**

Tipe Data Python dengan salah satu dari dua nilai bawaan, Benar atau Salah. Objek Boolean yang sama dengan True adalah benar (true), dan objek yang sama dengan False adalah salah (salah). Namun objek non-Boolean juga dapat dievaluasi dalam konteks Boolean dan ditentukan apakah benar atau salah. Hal ini dilambangkan dengan kelas bool.



```python
benar = True

salah = False

print(benar)

print(salah)

print(type(benar))

print(type(salah))
```

5. **Set**

Dalam Tipe Data Python, Set adalah kumpulan tipe data tidak berurutan yang dapat diubah, dan tidak memiliki elemen duplikat. Urutan unsur-unsur dalam suatu himpunan tidak terdefinisi meskipun dapat terdiri dari berbagai unsur.


```python
ini_set = set("DuniaCoding")

print(ini_set)

ini_set = set(["Dunia", "Coding", "Danial"])

print(ini_set)

print(type(ini_set))
```

6. Dictionary

Dictionary dengan Python adalah kumpulan nilai data yang tidak berurutan, digunakan untuk menyimpan nilai data seperti map, tidak seperti Tipe Data Python lainnya yang hanya menyimpan satu nilai sebagai elemen, Dictionary menyimpan kunci: pasangan nilai. Nilai kunci disediakan dalam Dictionary agar lebih optimal. Setiap pasangan kunci-nilai dalam Dictionary dipisahkan dengan titik dua : , sedangkan setiap kunci dipisahkan dengan 'koma'.

```python
ini_dictionary = {
    0 : "tahu bulat",
    1 : "Dunia",
    2 : "Codng",
    3 : 3,
    "nama": "Danial"
}

print(ini_dictionary)

print(type(ini_dictionary))

print(ini_dictionary["nama"])
```

1. if Statement (if... elif... else)
- pernyataan if adalah pernyataan pengambilan keputusan yang paling sederhana. Jika kondisi bernilai True, blok kode di dalam pernyataan if akan dieksekusi.

- Pernyataan if-elif-else dengan Python digunakan untuk pengambilan keputusan multi-arah. Hal ini memungkinkan Anda memeriksa beberapa kondisi secara berurutan dan mengeksekusi blok kode tertentu ketika suatu kondisi bernilai Benar.
```python
a = 1

if (a == '1'):
  print("Halo Dunia Coding")
elif (a == 1):
  print("Halo semuanya")
else:
  print("Halo Danial")
```
